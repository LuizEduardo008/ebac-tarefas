Nestes exercícios, você deve decidir qual é o gráfico visto em aula que melhor visualiza uma base de dados. Após decidir, você deverá criar a visualização usando o conteúdo exposto durante a aula e adicionar um pequeno parágrafo sobre um insights que pode ser extraido do gráfico.

## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
diamonds = sns.relplot(data=data,x="carat", y="price", hue="cut", style="cut")
```

```
diamonds.ax.set_title("Diamantes", fontsize=12, fontweight="bold");
diamonds.set_xlabels("Preço (USD)", fontsize=10);
diamonds.set_ylabels("Corte ", fontsize=10);
```

```
altura  = 25 / 2.54
largura = 25 / 2.54

diamonds.fig.set_size_inches(w=largura, h=altura)
```

```
diamonds.fig
```

**Insight do gráfico 1**: ...

O gráfico 1 possui muitos dados portanto sua visualização é mais complexa, mas basicamente foi separado no eixo x os tipos de Corte, no eixo y os preços de cada corte de diamante com a legenda para os tipos de corte.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights = sns.relplot(data=data,x="passengers", y="year", hue="month", style="month")
```

```
flights.ax.set_title("Voos", fontsize=12, fontweight="bold");
flights.set_xlabels("Passageiros", fontsize=10);
flights.set_ylabels("Ano", fontsize=10);
```

```
altura  = 20 / 2.54
largura = 20 / 2.54

flights.fig.set_size_inches(w=largura, h=altura)
```

```
flights.fig
```

**Insight do gráfico 2**: ...

O gráfico 2 fica um pouco melhor de se visualizar, por exemplo dá para notar que os meses de junho e julho são os que tem mais mais passageiros desde 1950, o gráfico tem o eixo x que são os anos e eixo y número de passageiros com a legenda dos meses atribuindo um símbolo para diferenciar.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")
  grafico.set(title='Passageiros por mês entre 1949 e 1959', xlabel='Mês', ylabel='Passageiros');
  grafico.get_legend().set_title("Ano");
```

**Insight do gráfico 3**: ...

O gráfico 3 é quase a mesma ideia do 2 só que dessa vez a informação que buscamos está entre os anos 1949 e 1959 em modelo de melhor visibilidade dos dados neste caso o eixo x se tornou o número de passageiros e o mês eixo y com os anos na legenda e pra cada um deles uma cor para localizarmos no gráfico criado.
